```python
url = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2019-09.csv.gz'


parse_dates = ['lpep_pickup_datetime','lpep_dropoff_datetime']
df_green = pd.read_csv(url, sep = ",", compression = "gzip", parse_dates=parse_dates, low_memory=False)
df_green
```

```python
df_green.dtypes
```

```python
start_date = '2019-09-18 00:00:00'
end_date = '2019-09-18 23:59:59'

filtered_df = df_green[(df_green['lpep_pickup_datetime'] >= start_date) & (df_green['lpep_dropoff_datetime'] <= end_date)]

# Imprimir el DataFrame resultante
print(filtered_df)
```

```python
# Convert lpep_pickup_datetime to datetime
df_green['lpep_pickup_datetime'] = pd.to_datetime(df_green['lpep_pickup_datetime'])

# Extract the day of the week
df_green['pickup_day'] = df_green['lpep_pickup_datetime'].dt.date

# Find the day with the largest total trip distance
max_day = df_green.loc[df_green['trip_distance'].idxmax(), 'pickup_day']

print(f"The pick-up day with the largest trip distance is: {max_day}")
```

```python
url = 'https://s3.amazonaws.com/nyc-tlc/misc/taxi+_zone_lookup.csv'


parse_dates = ['lpep_pickup_datetime','lpep_dropoff_datetime']
df_zones = pd.read_csv(url, sep = ",", low_memory=False)
df_zones
```

```python

# 1. Unir los dos DataFrames
df_merged = pd.merge(df_green, df_zones, left_on='PULocationID', right_on='LocationID')

# 2. Filtrar datos para 'lpep_pickup_datetime' en '2019-09-18' y Borough no igual a 'Unknown'
filtered_data = df_merged[(df_merged['lpep_pickup_datetime'].dt.date == pd.to_datetime('2019-09-18').date()) & (df_merged['Borough'] != 'Unknown')]

# 3. Agrupar por Borough y calcular la suma de total_amount
borough_totals = filtered_data.groupby('Borough')['total_amount'].sum()

# 4. Seleccionar los 3 Boroughs con la suma de total_amount superior a 50000
top_boroughs = borough_totals[borough_totals > 50000].nlargest(3)

print("Los 3 Boroughs con la suma de total_amount superior a 50000 son:")
print(top_boroughs)
```

```python
# 1. Unir los dos DataFrames basados en PULocationID y DOLocationID
df_merged = pd.merge(df_green, df_zones, left_on='PULocationID', right_on='LocationID', how='inner')

# 2. Filtrar datos para pasajeros recogidos en septiembre de 2019 en la zona Astoria
astoria_data = df_merged[(df_merged['lpep_pickup_datetime'].dt.year == 2019) & 
                         (df_merged['lpep_pickup_datetime'].dt.month == 9) & 
                         (df_merged['Zone'] == 'Astoria')]

# 3. Encontrar la zona de destino con la propina más grande
max_tip_zone = astoria_data.loc[astoria_data['tip_amount'].idxmax(), 'DOLocationID']

# 4. Obtener el nombre de la zona correspondiente al LocationID
max_tip_zone_name = df_zones.loc[df_zones['LocationID'] == max_tip_zone, 'Zone'].values[0]

print(f"Para los pasajeros recogidos en Astoria en septiembre de 2019, la zona de destino con la propina más grande es: {max_tip_zone_name}")
```
